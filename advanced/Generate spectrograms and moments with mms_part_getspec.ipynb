Get started by importing `mms_part_getspec`

```python
from pyspedas.mms.particles.mms_part_getspec import mms_part_getspec
from pytplot import tplot
```

### FPI spectrograms and moments

Notes: 
- as with the load routines, most keywords have defaults, e.g., the instrument keyword defaults to 'fpi'
- internal and external photoelectron corrections are applied to FPI electrons prior to generating the moments

```python
mms_part_getspec(trange=['2015-10-16/13:05:40', '2015-10-16/13:06:10'], 
                 instrument='fpi',
                 data_rate='brst', 
                 probe=1, 
                 species='i', 
                 output=['phi', 'theta', 'energy', 'pa', 'gyro', 'moments'])
```

```python
tplot(['mms1_dis_dist_brst_energy', 'mms1_dis_dist_brst_density'])
```

### HPCA spectrograms and moments

```python
mms_part_getspec(trange=['2015-10-16/10:00', '2015-10-16/11:00'], 
                 instrument='hpca',
                 probe=1, 
                 species='hplus', 
                 output=['phi', 'theta', 'energy', 'pa', 'gyro', 'moments'])
```

```python
tplot(['mms1_hpca_hplus_phase_space_density_energy', 
       'mms1_hpca_hplus_phase_space_density_density'])
```
