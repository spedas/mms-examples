This notebook shows how to create a spectra plot with EIS and HPCA data on the same panel


**** WARNING ON USING THIS NOTEBOOK ****

The EPD and HPCA instrument teams have not fully investigated the 
cross-calibration of these instruments; combining these spectra 
should be done on a case-by-case basis. Please contact the instrument 
teams before combining data from the EPD and plasma instruments.

```python
import pyspedas
```

Load the HPCA data

```python
pyspedas.mms.hpca(trange=['2015-10-16', '2015-10-17'], datatype='ion', probe=1)
```

```python
from pyspedas import mms_hpca_calc_anodes, mms_hpca_spin_sum
```

```python
mms_hpca_calc_anodes(fov=[0, 360])

mms_hpca_spin_sum(probe='1')
```

```python
pyspedas.mms.eis(trange=['2015-10-16', '2015-10-17'], datatype='extof', probe=1)
```

convert the EIS energies to eV so the 2 data products are on the same scale

```python
from pytplot import get_data, store_data
```

```python
data = get_data('mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin')
metadata = get_data('mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin', metadata=True)
```

```python
data.v
```

```python
store_data('mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin', data={'x': data.times, 'y': data.y, 'v': data.v*1000.0}, attr_dict=metadata)
```

```python
from pytplot import tplot
```

```python
tplot('mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin')
```

```python
store_data('combined_flux', data='mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin mms1_hpca_hplus_flux_elev_0-360_spin')
```

Set plot options

```python
from pytplot import options
```

```python
options('mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin', 'zrange', [0.1, 1e7])
options('mms1_hpca_hplus_flux_elev_0-360_spin', 'zrange', [0.1, 1e7])
options('mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin', 'yrange', [1, 1e7])
options('mms1_hpca_hplus_flux_elev_0-360_spin', 'yrange', [1, 1e7])
options('mms1_epd_eis_srvy_l2_extof_proton_flux_omni_spin', 'ztitle', '')
options('combined_flux', 'ytitle', 'EIS - HPCA')
options('combined_flux', 'ysubtitle', '[eV]')
```

```python
tplot('combined_flux')
```
