To access the data (and time) values from a tplot variable, use `get_data`:

```python
from pytplot import get_data

data = get_data('mms1_mec_r_sm')
```

The time values (unix times, UTC) are stored in `data.times`

```python
data.times
```

We have routines for converting unix times to/from time strings, and datetime objects, e.g.,

```python
from pyspedas import time_string
```

```python
time_string(data.times[0:5])
```

To change the format of the output string, use the fmt option, e.g.,

Note: see the strftime cheatsheet at https://strftime.org for possible formats

```python
time_string(data.times[0:5], fmt='%B %-d, %Y at %-H:%M:%S')
```

To convert back to unix times:

```python
from pyspedas import time_double
```

```python
time_double(['2015-10-16 00:00:00.000000',
             '2015-10-16 00:00:30.000000',
             '2015-10-16 00:01:00.000000',
             '2015-10-16 00:01:30.000000',
             '2015-10-16 00:02:00.000000'])
```

To convert to `datetime` objects, use `time_datetime`, e.g., 

```python
from pyspedas import time_datetime
```

If you work with `datetime` objects, be sure to set the time zone to UTC!

```python
time_datetime(data.times[0:5])
```

and the data values are stored in `data.y`

```python
data.y
```

```python
metadata = get_data('mms1_mec_r_sm', metadata=True)
```

```python
metadata
```

To create tplot variables, use `store_data`

```python
from pytplot import store_data

store_data('my_tplot_variable', data={'x': data.times, 'y': data.y})
```

```python
tplot('my_tplot_variable')
```

To preserve the metadata, you can also pass the dictionary returned by `get_data` to `store_data`

```python
store_data('my_tplot_variable', data={'x': data.times, 'y': data.y}, attr_dict=metadata)
```

Since we supplied the metadata dictionary, the plot options (legend, y-axis title, units in the subtitle, colors, etc.) should all be there

```python
tplot('my_tplot_variable')
```

Just like in IDL, we have a tool for easily converting tplot variables from Km to Re (and vice versa), `tkm2re`

```python
from pyspedas import tkm2re

tkm2re('my_tplot_variable')
```

Now we can plot the position data in Re instead of km. 

Note: `tkm2re` even updates the units in the y-axis subtitle

```python
tplot('my_tplot_variable_re')
```

## FIELDS Data

Load some burst-mode Fluxgate Magnetometer (FGM) data for all 4 probes

Note: here, we're time clipping the data to the requested time range; if you don't set `time_clip` to `True`, you might get more data than you requested. 

```python
pyspedas.mms.fgm(trange=['2016-11-23/07:49:32', '2016-11-23/07:49:35'], 
                 data_rate='brst', 
                 probe=[1, 2, 3, 4], 
                 time_clip=True)
```

and some burst-mode MEC data (GSE coordinates) for all 4 probes

```python
pyspedas.mms.mec(trange=['2016-11-23/07:49:32', '2016-11-23/07:49:35'], data_rate='brst', probe=[1, 2, 3, 4], time_clip=True, varformat='*_?_gse')
```

Now we can plot the FGM data, for all 4 probes:

```python
tplot('mms?_fgm_b_gse_brst_l2_bvec')
```

Now we can do curlometer calculations

```python
from pyspedas.mms import mms_curl

mms_curl(fields=['mms1_fgm_b_gse_brst_l2_bvec', 'mms2_fgm_b_gse_brst_l2_bvec', 'mms3_fgm_b_gse_brst_l2_bvec', 'mms4_fgm_b_gse_brst_l2_bvec'],
         positions=['mms1_mec_r_gse', 'mms2_mec_r_gse', 'mms3_mec_r_gse', 'mms4_mec_r_gse'])
```

Note: another way to access this same routine is by importing `curlometer` from `pyspedas.mms`, e.g., 

```python
from pyspedas.mms import curlometer

curlometer(fields=['mms1_fgm_b_gse_brst_l2_bvec', 'mms2_fgm_b_gse_brst_l2_bvec', 'mms3_fgm_b_gse_brst_l2_bvec', 'mms4_fgm_b_gse_brst_l2_bvec'],
           positions=['mms1_mec_r_gse', 'mms2_mec_r_gse', 'mms3_mec_r_gse', 'mms4_mec_r_gse'])
```

Now we can add the total current from the curlometer technique to the tplot of the previous figure

```python
tplot(['jtotal',
       'mms1_fgm_b_gse_brst_l2_bvec',
       'mms2_fgm_b_gse_brst_l2_bvec',
       'mms3_fgm_b_gse_brst_l2_bvec',
       'mms4_fgm_b_gse_brst_l2_bvec'])
```

## Plasma (FPI/HPCA) Data

Load the ion and electron moments from FPI

Note: use the `center_measurement` keyword for FPI and HPCA to shift the times to the center of the accumulation interval!

```python
pyspedas.mms.fpi(center_measurement=True, 
                 datatype=['dis-moms', 'des-moms'], 
                 trange=['2016-11-23', '2016-11-24'], 
                 probe=4)
```

Note: the legends default to `matplotlib`'s `best` option; we recently added support for other options (such as a `spedas` option to move the legend to the right of the panel), but these are still being tested and aren't available quite yet

```python
tplot(['mms4_des_errorflags_fast_moms_flagbars_full',
       'mms4_des_numberdensity_fast',
       'mms4_dis_errorflags_fast_moms_flagbars_full',
       'mms4_dis_numberdensity_fast'])
```

Just like with IDL `tplot`, `pytplot` now supports error bars on variables!

To add error bars to a tplot variable:

```python
dis_n = get_data('mms4_dis_numberdensity_fast')
dis_metadata = get_data('mms4_dis_numberdensity_fast', metadata=True)
dis_err = get_data('mms4_dis_numberdensity_err_fast')
```

Note: we're multiplying the errors by 100 so that they're clear on the figure

```python
error_data = dis_err.y*100
```

Note: the error data is specified in the 'dy' item in the data dictionary

```python
store_data('dis_density_with_errs', data={'x': dis_n.times, 'y': dis_n.y, 'dy': error_data})
```

The density data now has error bars attached

```python
tplot('dis_density_with_errs')
```

Zoom in to see the error bars

```python
from pytplot import tlimit

tlimit(['2016-11-23 13:00:00', '2016-11-23 13:05:00'])
```

```python
tplot('dis_density_with_errs')
```

Reset the time range

```python
tlimit(full=True)
```
