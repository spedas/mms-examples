# MMS Orbit Plots

This notebook shows how to create plots of the MMS orbits

This command sets up the PySPEDAS environment when running in Google Colab.  You can skip it or comment it out if you're running this notebook in an environment that already has PySPEDAS installed.

```python
#!pip install pyspedas
```

```python
from pyspedas.projects.mms.mms_orbit_plot import mms_orbit_plot
```

## Create the plot for all 4 spacecraft for the entire day of Oct 16, 2015 in GSE coordinates

```python
mms_orbit_plot(trange=['2015-10-16', '2015-10-17'])
```

## Change the coordinate system to SM

```python
mms_orbit_plot(trange=['2015-10-16', '2015-10-17'], coord='sm')
```

## Change the plane to XZ

```python
mms_orbit_plot(trange=['2015-10-16', '2015-10-17'], plane='xz', xr=[-10, 10], yr=[-10, 10])
```
